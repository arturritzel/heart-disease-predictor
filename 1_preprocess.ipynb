# Pr Processamento


```python
import numpy as np
import pandas as pd

#df de dataframe
df = pd.read_csv('/content/drive/MyDrive/MLcP_udemy/heart_tratado.csv',
                        sep = ';', encoding = 'utf-8')

print(df.head())
print(df.shape)
```

## Transformando as variveis categricas nominais em variveis categricas ordinais

```python
df2 = pd.DataFrame.copy(df)

df2.Sex.replace({'M': 0, 'F': 1}, inplace = True)
df2.ChestPainType.replace({'TA': 0, 'ATA' : 1, 'NAP' : 2, 'ASY' : 3}, inplace = True)
df2.RestingECG.replace({'Normal' : 0, 'ST' : 1, 'LVH' : 2}, inplace = True)
df2.ExerciseAngina.replace({'N' : 0, 'Y': 1}, inplace = True)
df2.ST_Slope.replace({'Up' : 0, 'Flat' : 1, 'Down' : 2}, inplace = True)

df2.head()
```

## Atributos previsores e alvo

```python
# previsores: argumentos utilizados pra descobrir o alvo

previsores = df2.iloc[:, 0:11].values
# iloc: localiza / ":": todas as linhas
# 0:11": intervalo. na prtica vai de 0 at 10, tipo um for i < 11
# .values: formato de array/matriz

previsores
```

```python
# alvo/target/classe:
alvo = df2.iloc[:, 11].values
# quando nao tem intervalo, ele pega o argumento q tu pediu: 11

alvo
```

## Anlise das escalas dos atributos (Escalonamento)

```python
# Padronizao: utiliza a mdia e o desvio padro como referncia (vamos utilizar)
# Normalizao: utiliza os valores mximos e mnimos como referncia

from sklearn.preprocessing import StandardScaler # standard scaler = escala de padronizao

previsores_esc = StandardScaler().fit_transform(previsores)
previsores_esc
# agora tudo varia em torno de 0, 1
```

```python
previsoresdf = pd.DataFrame(previsores_esc)
```

```python
previsoresdf.describe()
```

## Codificao de variveis categricas automaticamente

```python
# LabelEncoder: transformao de variveis categricas em numricas
from sklearn.preprocessing import LabelEncoder

previsores2 = df.iloc[:, 0:11].values
previsores2
```

```python
previsores2[:, 1] = LabelEncoder().fit_transform(previsores2[:,  1])
previsores2[:, 2] = LabelEncoder().fit_transform(previsores2[:,  2])
previsores2[:, 6] = LabelEncoder().fit_transform(previsores2[:,  4])
previsores2[:, 8] = LabelEncoder().fit_transform(previsores2[:,  8])
previsores2[:,10] = LabelEncoder().fit_transform(previsores2[:, 10])
# transformando apenas as colunas categricas

previsores2
```

## OneHotEncoder: criao de variveis Dummy/fictcias

```python
# se 0 for azul, 1 for verde e 2 for vermelho. 2 pessoas gostam de vermelho e 2 gostam de azul.
# a mdia de cores seria verde? no. isso  uma categoria, no um valor.
# o algoritmo no pode entender como valores maiores ou menores, e sim como diferentes.
# pra isso, criamos uma espcie de sequncia, onde azul na verdade  0-0-1, verde  0-1-0 e vermelho  1-0-0
# ele entende como diferentes, mas no como maiores/menores.

# con: deixa a tabela feia kkkkkkkkkkkkkkkkkkk
# ele criaria 3 colunas no lugar de uma, nesse exemplo.

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# parametros ColumnTransformer:
## name: nome dado a transformacao
## transformer: tipo de transformador (vamos usar o OneHotEncoder)
## columns: colunas que vao ser transformadas
## remainder: qq acontece com as outras colunas nao relacionadas: drop(padro): exclui /// passthrough: mantm
## sparse_threshold: parametro de classificaao de matrizes esparsas. default 0.3
## n_jobs: nmero de trabalhos a serem executados em paralelo. default nenhum
## transformer_weights: definio de pesos ao transformadores.
## verbose: True se quiser q a execuo aparea na tela

previsores3 = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1,2,6,8,10])],
                                remainder = 'passthrough').fit_transform(previsores)

previsores3.shape
```

```python
previsores3
```

```python
previsores3df = pd.DataFrame(previsores3)
previsores3df.head()
## ainda falta o escalonamento
```

```python
from sklearn.preprocessing import StandardScaler
previsores3_esc = StandardScaler().fit_transform(previsores3)
previsores3_esc
```

```python
previsores3df = pd.DataFrame(previsores3_esc)
previsores3df
```

```python
previsores3df.describe()
```

## Reduo de dimensionalidade
o objetivo  selecionar os melhores componentes/atributos do algoritmo, atravs da anlise das variveis.

Anlise dos Componentes Principais (PCA)
- Seleo de caracterstica: seleciona os melhores atributos e utiliza sem transformaes. <- vamos ver
- Encontra relaes dos melhores atributos e cria novos atributos. <- no supervisionada

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4) # svd_solver = default = auto.
previsores_pca = pca.fit_transform(previsores2)

print(previsores_pca)
print(pca.explained_variance_ratio_) # quantos % cada variavel "explica"
print(sum(pca.explained_variance_ratio_)) # total
```

```python
# Kernel PCA
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=4, kernel='rbf')
previsores_kernel = kpca.fit_transform(previsores2)
print(previsores_kernel)
```

```python
# Anlise do Discriminante Linear (LDA):
# supervisionado, aplicada em situaes com muitos atributos previsores
# e tambm com o atributo alvo com muitas classes

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=1)
previsores_lda = lda.fit_transform(previsores2, alvo)
print(previsores_lda)
print(lda.explained_variance_ratio_)
```

# Base de Treino e Teste

parmetros:

arrays: nomes dos arrays previsores e alvo

test_size: tamanho em porcentagem dos dados de teste (o restante vai ser treino, ou vice versa se definir o train_size)


train_size: =================================  treino
_ _ _ _ _ _ default: none (mas calcula se definir algum)

random_state: nomeao de um estado aleatrio

shuffle: embaralhamento dos dados aleatorios, associado com o random_state

stratify: possibilidade de dividir os dados de forma estratificada, default  none: nesse caso  mantida a proporo de 30% de zeros e 70% de 1 no dataframe, por exemplo.




```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, alvo, test_size=0.3,
                                                        random_state = 0)

print(x_treino.shape)
print(x_teste.shape)
print(y_treino.shape)
print(y_teste.shape)
```

```python
y_teste # pra comparar com as previsoes
```

```python
# pra calcular a taxa de acerto dele
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_score(y_teste, previsoes_naive)
print("Acurcia: %.2f%%" % (accuracy_score(y_teste, previsoes_naive)*100))

# previsores     : 84.42
# previsores_esc : 84.42
# previsores2    : 84.06
# previsores3    : 84.78
# previsores3_esc: 84.78
```

```python
# era 0 e ele disse 0 | era x e ele disse y
# era y e ele disse x | era 1 e ele disse 1

confusion_matrix(y_teste, previsoes_naive)
```

```python
# precision: verdadeiros positivos / (verdadeiros positivos + falsos positivos)
# recall: taxa de identificacao correta
# f1-score: media da precisao e recall
# support: aparicoes na base de dados

print(classification_report(y_teste, previsoes_naive))
```

# Otimizao de Hiperparmetros

```python
from sklearn.metrics import roc_auc_score # utilizado na otimizao de hiperparametros
from sklearn.model_selection import GridSearchCV # metodo de otimizacao
from sklearn.ensemble import GradientBoostingClassifier # algoritmo novo pra brincar

gbm = GradientBoostingClassifier(random_state=0)

# determinando as modificacoes dos hiperparametros
param_grid = dict(
    n_estimators=[20,50,100,250],
    learning_rate=[0.05,0.1,0.5],
    max_depth=[1,2,3,4,5],
)

grid_search = GridSearchCV(gbm, param_grid, scoring='roc_auc', cv=4)
grid_search.fit(x_treino, y_treino)
grid_search.best_params_
```

```python
resultado = pd.DataFrame(grid_search.cv_results_)
resultado.sort_values(by='mean_test_score', ascending=False, inplace=True)
resultado[['param_max_depth', 'param_learning_rate','param_n_estimators','mean_test_score','std_test_score']].head()
```
